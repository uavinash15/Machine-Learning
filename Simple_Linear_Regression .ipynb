# **Simple Linear Regression**

 Predicting MBA Salary from Grade in 10th Marks



```python
import pandas as pd
import numpy as np
```

```python
mba_salary_df=pd.read_csv('MBA Salary.csv')
mba_salary_df
```

```python
mba_salary_df.head(10)
```

```python
mba_salary_df.info()
```

```python
import statsmodels.api as sm
X=sm.add_constant(mba_salary_df['Percentage in Grade 10']) # adds a new column with constand value
print(X)
```

```python
X.head(5)
```

```python
Y=mba_salary_df['Salary']
```

Splitting the Dataset into Training and Validation Sets


```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

# skleran=scikit learn
```

```python
train_X,test_X,train_Y,test_Y=train_test_split(X,
                                              Y,
                                              train_size=0.8,
                                              random_state=100)
```

```python
import statsmodels.api as sm
```

```python
mba_salary_lm=sm.OLS(train_Y,train_X).fit() # x is Independent values, # Y= dependent values
```

Printing Estimated Parameters and Interpreting Them


```python
print(mba_salary_lm.params)
```
