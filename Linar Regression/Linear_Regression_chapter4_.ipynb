# **Residual Analysis**

**Probability - Probability Plot(P-P Plot)**

```python
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
```

```python
mba_salary_resid=mba_salary_lm.resid
probplot=sm.ProbPlot(mba_salary_resid)
plt.figure(figsize=(10,10))
probplot.ppplot(line='45')
plt.title("P-P Plot of Regression Standardized Residual")
plt.show()
```

```python
sm.qqplot(mba_salary_resid, line='r')
plt.title('Q-Q Plot of Regression Standardized Residuals')
plt.show()
```

**Test of Homoscedasticity**

```python
def get_standardized_values(vals):
  return (vals - vals.mean()) / vals.std()
```

```python
mba_salary_lm.fittedvalues
```

```python
plt.scatter(get_standardized_values(mba_salary_lm.fittedvalues),get_standardized_values(mba_salary_resid))
plt.title("Residual plot - MBA Salary Prediction")
plt.xlabel("standardized predicted value")
plt.ylabel("standardized residuals")
plt.show()
```

# **Outlier Analysis**

**Z-Score**

```python
from scipy.stats import zscore
```

```python
mba_salary_df['z_score_salary']=zscore(mba_salary_df.Salary)
```

```python
mba_salary_df
```

```python
mba_salary_df[(mba_salary_df.z_score_salary>3)|(mba_salary_df.z_score_salary<-3)]
```

```python
# there are no outliers since no z score is greater than 3 and less than -3
```

why we flagged 3 as outlier?:Here's the reason:

Normal Distribution: Z-scores are based on the standard normal distribution. In a standard normal distribution, about 99.7% of the data falls within three standard deviations of the mean.
Statistical Significance: This means that a data point with a Z-score greater than 3 or less than -3 is more than three standard deviations away from the mean. Such a data point is statistically very unlikely to occur if the data were truly drawn from a normal distribution.
Convention: While there's no strict rule, a Z-score of 3 is a commonly used convention in statistics to identify potential outliers. It represents a point that is quite far from the bulk of the data.
It's important to remember that using a Z-score of 3 is a guideline, not a rigid rule. The appropriate threshold for outlier detection can sometimes depend on the specific dataset and the context of your analysis. However, for many datasets, a Z-score of 3 is a reasonable starting point for flagging values that might be outliers.

In your case, since no data points had a Z-score greater than 3 or less than -3 in cell -pSaEOO54W2Z, it suggests there are no extreme outliers based on this criterion.

# **Cook's Distance**

```python
import numpy as np
mba_influence=mba_salary_lm.get_influence()
#print(mba_influence)
(c,p)=mba_influence.cooks_distance
#print((c,p))
plt.stem(np.arange(len(train_X)),np.round(c,3),markerfmt=",")
plt.title("Cooks distance for all observations in MBA Salary data set")
plt.xlabel("Row Index")
plt.ylabel("cooks's Distance")
plt.show()
```

**Calculating Prediction Intervals**

```python
# WLS: Weighted least squares
# wls_prediction_std=> for standard errors,confidence Intervals, prediction Intervals
```

```python
from statsmodels.sandbox.regression.predstd import wls_prediction_std

#predstd=> prediction standard deviation
pred_y=mba_salary_lm.predict(test_X)
std,pred_y_low,pred_y_high=wls_prediction_std(mba_salary_lm,
                                            test_X,
                                            alpha=0.1)
'''
print('std        : ',std)
print('pred_y_low :',pred_y_low)
print('pred_y_high:',pred_y_high)
'''

pred_y_df=pd.DataFrame({
    'grade_10_perc': test_X['Percentage in Grade 10'],
    'pred_y': pred_y,
    'pred_y_left': pred_y_low,
    'pred_y_right': pred_y_high})
```
